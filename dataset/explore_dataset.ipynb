### 15 classes

```python
label = raw_label.iloc[:,1:]
label.head()
```

```python
label.shape
```

```python
total_No_label_count = 0

for i in range(len(label)):
    if sum(label.iloc[i].values)==0:
        total_No_label_count += 1
```

```python
total_No_label_count
```

```python
count_of_disease = {}

for name in list(label.columns.values):
    count_of_disease[name] = sum(label[name].values)
   
count_of_disease["No label"] = total_No_label_count
count_of_disease
```

```python
plt.figure(figsize=(20,10))
plt.bar(*zip(*count_of_disease.items()))
plt.show()
```
